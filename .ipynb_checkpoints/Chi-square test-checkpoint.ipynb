```python
import numpy as np
import pandas as pd
import scipy.stats as stats 
from scipy.stats import chi2_contingency
from sklearn.feature_selection import SelectKBest 
from sklearn.feature_selection import chi2 
import matplotlib.pyplot as plt 
```

```python
cph_listings_df = pd.read_csv("data/cph_listings_df_clean.csv")
```

## Chi-squared testing

### Hypotheses Testing

- Ho (null hypothesis): Y does not depend on X or vice versa
- Ha (alternative hypothesis ): Variables are dependent

```python
# find observations for each combination
contingency_table = pd.crosstab(cph_listings_df['host_is_superhost'], cph_listings_df['instant_bookable'])
print(contingency_table)
```

```python
chi2, p, dof, expected = chi2_contingency(contingency_table)
print("Degrees of freemdom:", dof)
print("Chi-square statistic:", chi2)
print("P-value:", p)
```

- Chi-sqaure: deviation between observed and predicted
- P-value: low p-value indicates significant differences between observed and expected
- Dof: number of independent dataelements

```python
print(expected)
```

```python
# risk 5% 
alpha = 0.05
```

```python
# interpret p-value
if p <= alpha:
    print('variables are dependent, reject Ho')
else:
    print('Variables are independent, Ho holds true)')
```

### Visualisation

```python
# find critical value
critical_value = stats.chi2.ppf(1 - alpha, dof) 
critical_value
```

```python
chi2
```

```python
# generate values for the chi-square distribution 
x = np.linspace(0, 30, 1000) # from 0 - 30 with 1000 points
y = stats.chi2.pdf(x, dof)
  
# plotting
plt.figure(figsize=(10, 6))

plt.plot(x, y, label='Chi-Square Distribution (dof=1)') 
plt.fill_between(x, y, where=(x > critical_value), color='red', alpha=0.5, label='Critical Region') 
plt.axvline(chi2, color='blue', linestyle='dashed', label='Calculated Chi-Square') 
plt.axvline(critical_value, color='green', linestyle='dashed', label='Critical Value') 

plt.title('Chi-Square Distribution and Critical Region') 
plt.xlabel('Chi-Square Value') 
plt.ylabel('Probability Density Function') 
plt.legend() 
plt.grid(True)
plt.show()
```
